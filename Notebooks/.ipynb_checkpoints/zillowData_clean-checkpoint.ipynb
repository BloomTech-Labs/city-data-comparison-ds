```python
def merge_names(X):
    
    full_name = []
    for x in range(0,len(X.RegionName)):
        full_name.append(X.RegionName[x] + ' ' + X.State[x])
    
    X.columns
    
    #insert full_name column
    X['full_name'] = full_name
    
    # rearrange columns (dropped state and county name)
    X = X[cols]
    
    return X    
```
