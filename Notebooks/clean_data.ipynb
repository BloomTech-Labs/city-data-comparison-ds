# Cleaning the data

### In this notebook we clean the data in the all_census_data.csv file

```python
# imports
import pandas as pd
import numpy as np

df = pd.read_csv('all_census_data.csv');
```

```python
# Clean Data

def clean(x):
    if type(x) == str:
        if '-' in x:
            return 0
        elif '(X)' in x:
            return np.nan
        #otherwise...
    return x


def wrangle(X):
    """
    This function will clean and restructure the dataset
    """
    X = X.drop(index = 0) # drop first row
    
    for column_name in df.columns:
        X[column_name] = X[column_name].apply(clean)
        
    return X
```

```python
clean_df = wrangle(df)
```
