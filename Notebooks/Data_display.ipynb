```python
nm = 2312
jsn = {'_id':nm,
       'name':df.NAME[nm],
       'Marital Status':{'Males':{'Never Married':df['DP02_0025PE'][nm],
                                  'Currently Married':df['DP02_0026PE'][nm], 
                                  'Separated':df['DP02_0027PE'][nm], 
                                  'Divorced':df['DP02_0028PE'][nm]},
                        'Females':{'Never Married':df['DP02_0031PE'][nm],
                                  'Currently Married':df['DP02_0032PE'][nm], 
                                  'Separated':df['DP02_0033PE'][nm], 
                                  'Divorced':df['DP02_0034PE'][nm]}
                        },
       'Recent Mothers':{'Total':df['DP02_0036E'][nm],
                        'Birth Rate':{'Avg':df['DP02_0039E'][nm]},
                        'Age Distribution':{'15-19':df['DP02_0040E'][nm],
                                           '20-34':df['DP02_0041E'][nm],
                                           '35-50':df['DP02_0042E'][nm]},
                        'National Avg.':{'15-19':17.29,
                                           '20-34':103.95,
                                           '35-50':19.82}}
      }

jsn
```

```python
list(jsn['Recent Mothers']['Age Distribution'].keys())
```

```python
import plotly.graph_objects as go
import numpy as np

marrage_stats = go.Figure(data=[
    go.Bar(name='Males', x=list(jsn['Marital Status']['Males'].keys()), y=list(jsn['Marital Status']['Males'].values())),
    go.Bar(name='Females', x=list(jsn['Marital Status']['Males'].keys()), y=list(jsn['Marital Status']['Males'].values()))
])

marrage_stats.update_layout(barmode='group',title='Marriage Status')
marrage_stats.show()
x = np.arange(10)

print(f'{jsn["name"]} had a total of {jsn["Recent Mothers"]["Total"]} births in the last 12 months.')

births = go.Figure(data=[
    go.Bar(name=jsn['name'], x=list(jsn['Recent Mothers']['Age Distribution'].keys()), y=list(jsn['Recent Mothers']['Age Distribution'].values())),
    go.Bar(name='National Avg.', x=list(jsn['Recent Mothers']['National Avg.'].keys()), y=list(jsn['Recent Mothers']['National Avg.'].values()))
])
births.update_layout(barmode='group',
                     title='Births in the last 12 months per 1,000 women',
                     xaxis_title='Age Distribution')
births.show()
```
