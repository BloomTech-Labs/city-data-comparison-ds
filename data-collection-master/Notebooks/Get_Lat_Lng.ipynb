```python
def latcheck(ct):
    try:
        latlonmapper[ct[0]]
        res = True
    except:
        apilatlongdic[ct[1]]
        res = True
    else:
        res = False
    return(res)

def latplace(ct):
    try:
        loc = latlonmapper[ct[0]]
        lat = loc['lat']
        lng = loc['lng']
    except:
        loc = apilatlongdic[ct[1]]
        lat = loc['lat']
        lng = loc['lng']
    return((lat,lng))
```

```python
n = []
for i in df[['full_name', 'GEO_ID']].values:
    res = latcheck(i)
    n.append(res)
```

```python
lat = []
lng = []
for i in df[['full_name', 'GEO_ID']].values:
    res = latplace(i)
    lat.append(res[0])
    lng.append(res[1])
```

```python
df['lat'] = lat
df['lng'] = lng
```
