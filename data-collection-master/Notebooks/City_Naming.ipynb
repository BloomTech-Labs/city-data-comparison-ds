```python
import numpy as np
import pandas as pd
```

```python
uscb = pd.read_csv('all_census_data.csv')
```

```python
uscb = uscb.drop(index = 0) # drop first row
```

```python
uscb = uscb[['NAME', 'DP05_0001E']]
```

```python
# Make a new column with just the city name
uscb['City'] = uscb['NAME'].str.split(",", n = 1, expand=True)[0]

# Make a new column with just the state name
uscb['State'] = uscb['NAME'].str.split(",", n = 1, expand=True)[1]
uscb['State'] = uscb['State'].str.lstrip()

# Mark which type of 'place' each city is deamded to be by the US Census Bureau
uscb['Place_Type_city'] = uscb['City'].str.contains(pat = ' city')
uscb['Place_Type_town'] = uscb['City'].str.contains(pat = ' town')
uscb['Place_Type_borough'] = uscb['City'].str.contains(pat = ' borough')
uscb['Place_Type_CPD'] = uscb['City'].str.contains(pat = ' CDP')

# Remove that place type from the city name
uscb['City'] = uscb['City'].str.replace(' city', '', regex=False)
uscb['City'] = uscb['City'].str.replace(' town', '', regex=False)
uscb['City'] = uscb['City'].str.replace(' borough', '', regex=False)
uscb['City'] = uscb['City'].str.replace(' CDP', '', regex=False)
uscb['City'] = uscb['City'].str.rstrip()
```
