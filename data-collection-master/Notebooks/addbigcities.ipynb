```python
def makeframe(geoid,state,marker,st,full_name,has_loc,lat,lng):
    censvals = cens[cens.GEO_ID == geoid].values.tolist()[0]
    popdata = pop[pop.GEO_ID == geoid].values.tolist()[0][2:]
    cols = df.columns.tolist()
    li =[]
    for i in censvals:
        li.append(i)
    for i in popdata:
        li.append(i)
    maninfo = censvals + [state, marker, st, full_name] + popdata + [has_loc, lat, lng]

    data = pd.DataFrame([maninfo], columns=cols)
    return(data)
```
