```python
import pandas as pd
```

```python
census = pd.read_csv('data/census_df_zillow.csv')
```

```python
zillow_columns = ['full_name','zillow_house_bd1', 'zillow_house_bd2', 'zillow_house_bd3','zillow_house_bd4',
                  'zillow_house_bd5', 'zillow_rental_bd1','zillow_rental_bd2','zillow_rental_bd3','zillow_rental_bd4',
                  'zillow_rental_bd5']

zillow_check = pd.DataFrame(census, columns=zillow_columns)
```

```python
zillow_check.shape
```

```python
zillow_check.head()
```

```python
house_prices = ['zillow_house_bd1','zillow_house_bd2','zillow_house_bd3','zillow_house_bd4','zillow_house_bd5']

rental_prices = ['zillow_rental_bd1','zillow_rental_bd2','zillow_rental_bd3', 'zillow_rental_bd4','zillow_rental_bd5']
```

```python
housing_price_check = pd.DataFrame(zillow_check, columns=house_prices)
rental_price_check = pd.DataFrame(zillow_check, columns=rental_prices)

zillow_check['house_price_avail'] = housing_price_check.all(axis='columns')
zillow_check['rental_price_avail'] = rental_price_check.all(axis='columns')
```

```python
zillow_check.to_csv('data/zillow_price_avail.csv')
```
