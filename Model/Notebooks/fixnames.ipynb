## Dataset to help abbreviate each state

```python
copy = pd.read_html('https://www.infoplease.com/us/postal-information/state-abbreviations-and-state-postal-codes')
codes = copy[0]
zipped = dict(zip(codes['State/District'].values, codes['Postal Code'].values))
codes.head()
```

```python
import numpy as np
def abb(st):
    return(zipped[st])

df['st'] = df.state.apply(abb)
df['full_name'] = np.NaN * len(df)
df.head()
```

## Create a new dataset with formated city, town, village, and borough locations

```python
new = []
for i in df.values:
    check = i[4].split(',')[0]
    if ' city' in check:
        i[-1] = check.split(' city')[0].strip()
        i[-1] = f"{i[-1]} {i[-2]}"
        new.append(i)
    elif ' town' in check:
        i[-1] = check.split(' town')[0].strip()  
        i[-1] = f"{i[-1]} {i[-2]}"
        new.append(i)
    elif ' village' in check:
        i[-1] = check.split(' village')[0].strip() 
        i[-1] = f"{i[-1]} {i[-2]}"
        new.append(i)        
    elif ' borough' in check:
        i[-1] = check.split(' borough')[0].strip() 
        i[-1] = f"{i[-1]} {i[-2]}"
        new.append(i)
    elif check == 'Carson City':
        i[-1] = 'Carson City'  
        i[-1] = f"{i[-1]} {i[-2]}"
        new.append(i)
    elif check == 'Princeton':
        i[-1] = 'Princeton' 
        i[-1] = f"{i[-1]} {i[-2]}"
        new.append(i) 
    elif 'Nashville-Davidson metropolitan' in check:
        i[-1] = 'Nashville' 
        i[-1] = f"{i[-1]} {i[-2]}"
        new.append(i) 
    else:
        pass
    
```
